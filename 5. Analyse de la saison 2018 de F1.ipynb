### 2.1.2 Distrubition des variables météorologiques sur toute la saison

```python
def visualize_single_dimension_data(df):
    """
    Visualisation stylisée des données à une dimension pour chaque colonne numérique d'un DataFrame.
    
    Parameters:
    df (DataFrame) : Le DataFrame contenant les données numériques à visualiser.
    """
    # Définir un thème pour Seaborn
    sns.set_theme(style="whitegrid")
    
    # Palette de couleurs
    palette = sns.color_palette("viridis")
    
    # Sélectionner les colonnes numériques
    numeric_columns = df.select_dtypes(['int', 'float', 'timedelta64'])
    
    # Calculer le nombre de colonnes et de lignes nécessaires pour afficher tous les graphiques
    n_cols = 3  # Nombre de colonnes par ligne
    n_rows = (len(numeric_columns.columns) + n_cols - 1) // n_cols  # Nombre de lignes nécessaires
    
    # Créer une figure avec plusieurs sous-graphes
    fig, axes = plt.subplots(n_rows, n_cols, figsize=(18, n_rows * 4))
    axes = axes.flatten()  # Aplatir les axes pour un accès facile
    
    # Itérer sur chaque colonne numérique et afficher un graphique approprié
    for i, column in enumerate(numeric_columns.columns):
        ax = axes[i]

        sns.histplot(df[column], kde=True, color=palette[1], ax=ax, bins = 70)
        ax.set_title(f'Histogramme de {column}', fontsize=12, fontweight='bold')
        ax.set_xlabel(column, fontsize=10)
        
        # Ajuster le style des ticks
        ax.tick_params(axis='both', which='major', labelsize=9)
    
    # Supprimer les sous-graphes inutilisés
    for j in range(i + 1, len(axes)):
        fig.delaxes(axes[j])
    
    # Ajuster l'espacement entre les sous-graphes
    plt.tight_layout()
    
    # Afficher les graphiques
    plt.show()

columns = [ 'AirTemp', 'Humidity', 'Pressure', 'Rainfall', 'TrackTemp', 'WindDirection', 'WindSpeed']
visualize_single_dimension_data(meteo_2018[columns])
```

#### Interpretation 

- AirTemp (Température) :
La distribution est multimodale, avec plusieurs pics, mais ceux-ci ne sont pas identiques. Cela indique une répartition complexe des températures.

- Humidity (Humidité) :
La distribution est multimodale, légèrement symétrique vers la gauche, suggérant des variations dans les niveaux d'humidité avec une légère tendance vers des valeurs plus faibles.

- Pressure (Pression atmosphérique) :
La distribution est symétrique à gauche, avec une concentration des valeurs sur la gauche et quelques valeurs très faibles par rapport à la tendance générale.

- TrackTemp (Température de la piste) :
La distribution est symétrique à droite, avec quelques valeurs extrêmes élevées, mais elles sont peu fréquentes.

- WindDirection (Direction du vent) :
La distribution est presque symétrique par rapport au centre, avec une répartition quasi égale des valeurs à gauche et à droite, bien qu'il y ait une légère concentration vers la gauche.

- WindSpeed (Vitesse du vent) :
La distribution est homogène, avec quelques valeurs légèrement éloignées de la majorité.

### 2.1.3 Boxplot

```python
def plot_boxplots(df):
    """
    Visualisation stylisée des boxplots pour chaque colonne numérique d'un DataFrame.
    
    Parameters:
    df (DataFrame) : Le DataFrame contenant les données numériques pour lesquelles générer les boxplots.
    """
    # Définir un thème pour Seaborn
    sns.set_theme(style="whitegrid")
    
    # Sélectionner les colonnes numériques
    numeric_columns = df.select_dtypes(['int', 'float', 'timedelta64'])
    
    # Palette de couleurs
    palette = sns.color_palette("viridis")
    
    # Calculer le nombre de colonnes et de lignes nécessaires pour afficher tous les graphiques
    n_cols = 3  # Nombre de colonnes par ligne
    n_rows = (len(numeric_columns.columns) + n_cols - 1) // n_cols  # Nombre de lignes nécessaires
    
    # Créer une figure avec plusieurs sous-graphes
    fig, axes = plt.subplots(n_rows, n_cols, figsize=(18, n_rows * 4))
    axes = axes.flatten()  # Aplatir les axes pour un accès facile
    
    # Itérer sur chaque colonne numérique et afficher le boxplot
    for i, column in enumerate(numeric_columns.columns):
        ax = axes[i]
        sns.boxplot(data=df, x=column, ax=ax, color=palette[1], width=0.5, linewidth=1.5, showmeans=True)
        ax.set_title(f'Boxplot de {column}', fontsize=12, fontweight='bold')
        ax.set_xlabel(column, fontsize=10)
        
        # Améliorer les ticks pour lisibilité
        ax.tick_params(axis='both', which='major', labelsize=9)
    
    # Supprimer les sous-graphes inutilisés
    for j in range(i + 1, len(axes)):
        fig.delaxes(axes[j])
    
    # Ajuster l'espacement entre les sous-graphes
    plt.tight_layout()
    
    # Afficher les graphiques
    plt.show()

plot_boxplots(meteo_2018[columns])
```

La distribution est centrée autour de la moyenne et de la médiane, sans valeurs aberrantes. 50% des valeurs se concentrent entre 21 et 27,5°C, avec une dispersion modérée et une légère asymétrie vers la droite.

### 2.1.4 Conditions meteoroliques par course

```python
import seaborn as sns
import matplotlib.pyplot as plt
from windrose import WindroseAxes

def plot_weather_data_for_race(round_number, meteo_2018, courses_2018):
    # Filtrer les données de la course spécifiée
    course = courses_2018[courses_2018['round'] == round_number]
    if course.empty:
        print(f"Aucune course trouvée pour le Round Number {round_number}.")
        return

    race_name = course.iloc[0]['name']
    df_weather = meteo_2018[meteo_2018['Round Number'] == round_number]

    if df_weather.empty:
        print(f"Aucune donnée météo trouvée pour la course {race_name}.")
        return

    # Convertir le temps en minutes
    df_weather['TimeMinutes'] = df_weather['Time'].dt.total_seconds() / 60

    # Créer les sous-graphiques
    fig, ax = plt.subplots(2, 2, figsize=(15, 15))
    fig.suptitle(f'Weather Data & Track Evolution \n{race_name}', fontsize=30)

    # Températures de la piste et de l'air
    sns.lineplot(data=df_weather, x='TimeMinutes', y='TrackTemp', label='TrackTemp', ax=ax[0, 0])
    sns.lineplot(data=df_weather, x='TimeMinutes', y='AirTemp', label='AirTemp', ax=ax[0, 0])
    
    # Ajouter les périodes de pluie
    if 'Rainfall' in df_weather.columns and df_weather['Rainfall'].dtype == 'bool':
        rain_periods = df_weather[df_weather['Rainfall'] == True]
        if not rain_periods.empty:
            ax[0, 0].fill_between(
                rain_periods['TimeMinutes'], 
                df_weather['TrackTemp'].max() + 0.5, 
                df_weather['AirTemp'].min() - 0.5, 
                facecolor="blue", color='blue', alpha=0.1, zorder=0, label='Rain'
            )

    ax[0, 0].legend(loc='upper right')
    ax[0, 0].set_ylabel('Temperature (°C)')
    ax[0, 0].title.set_text('Track Temperature & Air Temperature')
    ax[0, 0].set_xlabel('Time (minutes)')

    # Humidité
    sns.lineplot(data=df_weather, x='TimeMinutes', y='Humidity', ax=ax[0, 1])
    ax[0, 1].title.set_text('Track Humidity (%)')
    ax[0, 1].set_xlabel('Time (minutes)')

    # Pression atmosphérique
    sns.lineplot(data=df_weather, x='TimeMinutes', y='Pressure', ax=ax[1, 0])
    ax[1, 0].title.set_text('Air Pressure (mbar)')
    ax[1, 0].set_xlabel('Time (minutes)')

    # Direction et vitesse du vent
    rect = ax[1, 1].get_position()
    wax = WindroseAxes(fig, rect)
    fig.add_axes(wax)
    wax.bar(df_weather['WindDirection'], df_weather['WindSpeed'], normed=True, opening=0.8, edgecolor='white')
    wax.set_legend()
    ax[1, 1].title.set_text('Wind Direction (°) and Speed (m/s)')

    plt.show()

# Exemple d'utilisation
plot_weather_data_for_race(
    round_number=1, 
    meteo_2018=meteo_2018, 
    courses_2018=courses_2018
)
```

Cette analyse permet de visualiser comment les conditions météorologiques varient au cours des différentes courses de la saison 2018 de Formula 1, minute par minute. À travers des graphiques représentant l'évolution des températures de la piste et de l'air, de l'humidité, de la pression atmosphérique, la pluie, ainsi que de la direction et de la vitesse du vent, nous observons les fluctuations de ces variables au fil du temps pendant chaque course

### 2.1.5 Variance des variables météorologiques pour chaque course

```python
# Calculer la moyenne et la variance des variables météorologiques par course
weather_stats = meteo_2018.groupby('Round Number').agg(
    AirTemp_variance=('AirTemp', 'var'),
    Humidity_variance=('Humidity', 'var'),
    Pressure_variance=('Pressure', 'var'),
    TrackTemp_variance=('TrackTemp', 'var'),
    WindDirection_variance=('WindDirection', 'var'),
    WindSpeed_variance=('WindSpeed', 'var')
)

weather_stats
```

#### Interprétation des variances des variables météorologiques :
L'analyse des variances des variables météorologiques nous permet de déterminer si nous pouvons représenter les conditions climatiques d'une course par leurs moyennes. Étant donné que les données sont mesurées minute par minute pendant les courses, nous cherchons à estimer les moyennes des variables météorologiques tout en nous assurant qu'elles ne varient que très peu durant la course. Les variables dont la variance est faible sont celles qui peuvent être représentées de manière fiable par leur moyenne, simplifiant ainsi l'analyse et les traitements ultérieurs.

#### Variables à considérer pour une représentation par la moyenne :

- Pressure (Pression) : La variance est faible dans la plupart des courses, ce qui permet de considérer la moyenne comme une bonne approximation de la pression durant la course. Cette variable est donc représentée de manière fiable par sa moyenne.

- AirTemp (Température de l'air) et TrackTemp (Température de la piste) : Ces variables présentent une variance modérée, mais leur moyenne reste relativement stable, surtout pour les courses où les variations sont faibles. Elles peuvent donc être simplifiées par leur moyenne dans la plupart des cas, à condition de vérifier les variations dans les courses à forte fluctuation.

- WindSpeed (Vitesse du vent) : Pour certaines courses avec une faible variance, la moyenne peut également être significative et utilisée pour représenter les conditions de vent. Néanmoins, pour des courses avec une forte variation de la vitesse du vent, des analyses plus approfondies peuvent être nécessaires.

#### Variables à analyser plus en détail :

- WindDirection (Direction du vent) et Humidity (Humidité) : Ces variables présentent une grande variance dans certaines courses, ce qui suggère que leur moyenne pourrait ne pas suffire pour capturer les variations significatives. Ces variables nécessitent des analyses plus fines pour comprendre pleinement l'impact de leurs variations pendant la course. Il est probable qu'elles ne soient pas représentées de manière adéquate uniquement par leur moyenne dans tous les cas.

```python
def plot_countplots(data, columns):
    """
    Fonction pour générer des countplots pour les colonnes spécifiées dans un dataset.

    Parameters:
    data (DataFrame): Le DataFrame contenant les données.
    columns (list): Liste des colonnes à afficher.

    """

    sns.set_style("whitegrid")  # Style de grille
    palette = sns.color_palette("viridis")  # Palette de couleurs

    # Créer un diagramme en barres pour chaque colonne
    fig, ax = plt.subplots(1, len(columns), figsize=(6 * len(columns), 6))

    for i, column in enumerate(columns):
        sns.countplot(x=data[column], ax=ax[i])
        ax[i].set_title(f'Distribution de {column}')
        ax[i].set_xlabel(column)
        ax[i].set_ylabel('Fréquence')

    # Ajuster l'espacement entre les graphiques
    plt.tight_layout()

    # Afficher le graphique
    plt.show()

columns = ['positionOrder', 'points', 'rank']
plot_countplots(results_2018, columns)
```

#### Interprétation :

- Distribution de positionOrder : La distribution est quasiment uniforme, ce qui signifie que les positions des pilotes sont réparties de manière assez égale à travers les différentes courses. Cela suggère qu'il n'y a pas de biais particulier dans l'attribution des positions parmi les pilotes, chaque position étant représentée de manière équivalente.

- Distribution de points : Le mode est très élevé à 0, ce qui montre qu'une grande majorité des pilotes ne marquent pas de points lors de chaque course. Les autres valeurs de points sont assez uniformes entre elles.
```python
def plot_time_distributions(data, columns):
    """
    Fonction pour générer des histogrammes avec KDE pour la distribution des temps des tours les plus rapides 
    et des temps de course, en fonction des colonnes spécifiées dans la liste.

    Parameters:
    data (DataFrame): Le DataFrame contenant les données.
    columns (list): Liste des noms des colonnes des temps de tour et des temps de course.
    """
    # Vérifier que la liste contient deux éléments pour les temps de tour et de course
    if len(columns) != 2:
        raise ValueError("La liste 'columns' doit contenir exactement deux éléments : le temps du tour et le temps de course.")

    # Extraire les noms des colonnes
    fastest_lap_column = columns[0]
    race_time_column = columns[1]
    
    # Configuration globale pour un style cohérent
    sns.set_style("whitegrid")  # Style de grille
    palette = sns.color_palette("viridis")  # Palette de couleurs

    # Créer une figure avec deux sous-graphiques (1 ligne, 2 colonnes)
    fig, axes = plt.subplots(1, 2, figsize=(16, 6))  # Taille ajustée pour une présentation optimale

    # Premier graphique : Distribution des Fastest Lap Times avec KDE
    sns.histplot(data[fastest_lap_column].dt.total_seconds(), kde=True, 
                 bins=60, color=palette[1], edgecolor='white', alpha=0.9, ax=axes[0])
    axes[0].set_xlabel('Fastest Lap Time (seconds)', fontsize=12)
    axes[0].set_ylabel('Frequency', fontsize=12)
    axes[0].set_title('Distribution of Fastest Lap Times', fontsize=14, fontweight='bold')
    axes[0].tick_params(axis='both', which='major', labelsize=10)

    # Deuxième graphique : Distribution des Time en heures avec KDE
    sns.histplot(data[race_time_column].dropna().dt.total_seconds() / 3600, kde=True, 
                 bins=70, color=palette[1], edgecolor='white', alpha=0.9, ax=axes[1])
    axes[1].set_xlabel('Time (hours)', fontsize=12)
    axes[1].set_ylabel('Frequency', fontsize=12)
    axes[1].set_title('Distribution of Time in Hours', fontsize=14, fontweight='bold')
    axes[1].tick_params(axis='both', which='major', labelsize=10)

    # Ajuster l'espacement entre les subplots pour éviter le chevauchement
    plt.tight_layout()

    # Afficher les graphiques
    plt.show()

# Appel à la fonction
columns = ['fastestLapTime_', 'time_']
plot_time_distributions(results_2018, columns)
```

#### Interpretation :

```python
# Convertir 'time_' (durée) en secondes pour une comparaison facile
results_with_meteo['time_seconds'] = results_with_meteo['time_'].dt.total_seconds()

# Agréger les données par raceId
aggregated_data = results_with_meteo.groupby('raceId').agg(
    time_=('time_seconds', 'min'),  # Garder le temps minimal
    fastestLapSpeed_=('fastestLapSpeed_', 'max'),  # Garder la vitesse maximale du meilleur tour
    AirTemp_avg=('AirTemp_avg', 'mean'),  # Moyenne de la température de l'air
    Pressure_avg=('Pressure_avg', 'mean'),  # Moyenne de la pression
    TrackTemp_avg=('TrackTemp_avg', 'mean'),  # Moyenne de la température de la piste
    WindSpeed_avg=('WindSpeed_avg', 'mean')  # Moyenne de la vitesse du vent
).reset_index()

# Convertir à nouveau 'time_' en timedelta (si vous voulez l'avoir sous cette forme)
aggregated_data['time_'] = pd.to_timedelta(aggregated_data['time_'], unit='s')

# Calculer la corrélation sur le dataset agrégé
correlation_matrix = aggregated_data.corr()

# Afficher la matrice de corrélation sous forme de heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Matrice de Corrélation')
plt.show()
```

#### 1. Variation de la vitesse vainqueur de chaque course en fonction variables meteorologiques

```python
# Fonction pour catégoriser la température de l'air
def categorize_air_temp(temp):
    if temp < 15:
        return 'Très froide'
    elif 15 <= temp < 25:
        return 'Moyenne'
    else:
        return 'Très chaude'

# Fonction pour catégoriser la pression
def categorize_pressure(pressure):
    if pressure < 980:
        return 'Très basse'
    elif 980 <= pressure < 1020:
        return 'Moyenne'
    else:
        return 'Très élevée'

# Fonction pour catégoriser la température de la piste
def categorize_track_temp(temp):
    if temp < 20:
        return 'Très froide'
    elif 20 <= temp < 30:
        return 'Moyenne'
    else:
        return 'Très chaude'
```

```python
# Regrouper les données par course (raceId) et conserver les moyennes des variables météorologiques
grouped_by_race = results_with_meteo.groupby('raceId').first().reset_index()

# Appliquer les fonctions de catégorisation aux colonnes correspondantes
grouped_by_race['AirTemp_cat'] = grouped_by_race['AirTemp_avg'].apply(categorize_air_temp)
grouped_by_race['Pressure_cat'] = grouped_by_race['Pressure_avg'].apply(categorize_pressure)
grouped_by_race['TrackTemp_cat'] = grouped_by_race['TrackTemp_avg'].apply(categorize_track_temp)

# Sélectionner les colonnes utiles pour le nouveau dataset
final_race_data = grouped_by_race[
    ['raceId', 'AirTemp_cat', 'Pressure_cat', 'TrackTemp_cat', 'Rainfall']
]

# Afficher un aperçu du nouveau dataset
final_race_data.head()
```

```python
# Calculer la vitesse maximale du tour le plus rapide pour chaque course
max_fastest_lap_speed = results_with_meteo.groupby('raceId')['fastestLapSpeed_'].max().reset_index()

# Fusionner avec le dataset final_race_data sur raceId
final_race_data = final_race_data.merge(max_fastest_lap_speed, on='raceId', how='left')

# Afficher un aperçu du dataset final
final_race_data.head()
```
